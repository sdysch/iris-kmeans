```python
# standard imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
data
```

```python
from sklearn.cluster import KMeans
X = data.drop(["target"], axis=1)
y = data["target"]
kmeans = KMeans(n_clusters=3, random_state=0)
y_kmeans = kmeans.fit(X)
y_pred = y_kmeans.labels_
```

```python
# plot features, colour by class
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X["sepal length (cm)"], X["sepal width (cm)"], c=y_pred, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(X["sepal length (cm)"], X["sepal width (cm)"], c=y, cmap='gist_rainbow', edgecolor='k', s=150)

axes[0].set_xlabel("Sepal length [cm]")
axes[0].set_ylabel("Sepal width [cm]")
axes[1].set_xlabel("Sepal length [cm]")
axes[1].set_ylabel("Sepal width [cm]")

axes[0].set_title("Prediction")
axes[1].set_title("Truth")
```

```python
# plot features, colour by class
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X["petal length (cm)"], X["petal width (cm)"], c=y_pred, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(X["petal length (cm)"], X["petal width (cm)"], c=y, cmap='gist_rainbow', edgecolor='k', s=150)

axes[0].set_xlabel("petal length [cm]")
axes[0].set_ylabel("petal width [cm]")
axes[1].set_xlabel("petal length [cm]")
axes[1].set_ylabel("petal width [cm]")

axes[0].set_title("Prediction")
axes[1].set_title("Truth")
```
